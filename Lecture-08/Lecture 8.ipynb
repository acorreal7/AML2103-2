```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

# Options for pandas
pd.options.display.max_columns = 100
pd.options.display.max_rows = 100

# Options for matplotlib
%matplotlib inline
plt.style.use('seaborn-v0_8')
mpl.rcParams['figure.dpi'] = 150

# Read the file
df = pd.read_csv('./CustomersTransactions2021.csv')
```

```python
# Assign subplots to variables ax1 ... ax6 and share ticks
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, sharey=True, sharex=True, figsize=(10, 5))

# 
def plot_customer_data(ax, customer_name, style):
    ax.plot(
        df[df['Customers'] == customer_name]["id"],
        df[df['Customers'] == customer_name]["txn_total"],
        style
    )
    ax.set_title(customer_name)

plot_customer_data(ax1, 'Johnny Awesome', "co--")
plot_customer_data(ax2, 'Bob Marley', "m.--")
plot_customer_data(ax3, 'Taylor Swift', "g*--")
plot_customer_data(ax4, 'Stephen Smith', "ro--")
plot_customer_data(ax5, 'Delilah Avery', "y+--")
plot_customer_data(ax6, 'Johnson Cory', "b*--")
```

#### 5 Bargraphs

```python
df['Category'].unique()
```

```python
df.groupby(by='Category').agg({ 'txn_total': 'sum' }).plot(kind='bar', width=0.7)
plt.title('Total txn_total per category')
plt.xlabel('Category')
plt.ylabel('Total')
plt.xticks(rotation=0)
plt.legend()
```

```python
sum_data = df.groupby(by='Category')['txn_total'].agg('sum').reset_index()

plt.bar(sum_data['Category'], sum_data['txn_total'], label='Category', width=0.7, color='c')
plt.title('Total txn_total per category')
plt.xlabel('Category')
plt.ylabel('Total')
plt.legend()
```
